## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 20, 3, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10))  # width, height

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10, 
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store",
       ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store", xlabel="Price ($)");
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)
x
```

```python
fig, ax = plt.subplots()
ax.hist(x);
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(20, 12))

# Plot to each different axis
ax1.plot(x, x/2)
ax1.set(title="Plot", xlabel="x", ylabel="y")
ax2.scatter(np.random.random(10), np.random.random(10))
ax2.set(title="Scatter", xlabel="random 10", ylabel="random 10")
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax3.set(title="Bar", xlabel="keys", ylabel="values")
ax4.hist(np.random.randn(1000));
ax4.set(title="Hist");
```

```python
# Subplots option 2

np.random.seed(0)
y = np.random.random(10)

plot_number = 4
ncols = 3

nrows = int(plot_number/ncols) + (1 if np.remainder(plot_number, ncols) != 0 else 0)
fig, xx = plt.subplots(ncols=ncols, nrows=nrows, figsize=(25, 12))
xx[0][0].plot(y, y**2)
xx[0][0].set(title="Plot", xlabel="y", ylabel="y**")
xx[0][1].scatter(y, y**2)
xx[0][1].set(title="Scatter", xlabel="y", ylabel="y**")
xx[0][2].bar(y, y**2)
xx[0][2].set(title="Bar", xlabel="y", ylabel="y**")
xx[1][0].barh(y, y**2)
xx[1][0].set(title="Bar H", xlabel="y", ylabel="y**")
xx[1][1].hist(y);
xx[1][1].set(title="Hist")
xx[1][2].hist(y, bins=20)
xx[1][2].set(title="Hist");
```

```python
# Make a dataframe
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace("[\$\,\.]", "", regex=True).str[:-2]
```

```python
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2025", periods=len(car_sales))
car_sales
```

```python
car_sales["Price"].astype(int).cumsum()
```

```python
pd.to_numeric(car_sales["Price"]).cumsum()
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# Let's plot the total sales
car_sales.plot("Sale Date");
```

```python
car_sales.plot(x="Sale Date",  y="Total Sales");
```
